```python
# Perceptron implementation for OR Gate

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def predict(self, inputs):
        # Calculate weighted sum
        total = 0
        for i in range(len(inputs)):
            total += inputs[i] * self.weights[i]
        total += self.bias

        # Step activation function
        if total >= 0:
            return 1
        else:
            return 0


# Suitable weights and bias for OR gate
weights = [1, 1]
bias = -0.5

# Create perceptron object
perceptron = Perceptron(weights, bias)

# OR gate input combinations
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

# Display output
print("X1 X2 Output")
for x in inputs:
    output = perceptron.predict(x)
    print(x[0], " ", x[1], "   ", output)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Perceptron implementation
class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def predict(self, inputs):
        total = 0
        for i in range(len(inputs)):
            total += inputs[i] * self.weights[i]
        total += self.bias

        if total >= 0:
            return 1
        else:
            return 0


# Suitable weights and bias for NOR gate
weights = [-1, -1]
bias = 0.5

# Create perceptron
perceptron = Perceptron(weights, bias)

# Input combinations
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

# Verify outputs
print("X1 X2 Output")
for x in inputs:
    output = perceptron.predict(x)
    print(x[0], " ", x[1], "   ", output)

# -----------------------------
# Draw Decision Boundary
# -----------------------------
plt.figure(figsize=(6,6))

# Plot points
for x in inputs:
    y = perceptron.predict(x)
    if y == 1:
        plt.scatter(x[0], x[1], color='green', s=100, label='Output = 1')
    else:
        plt.scatter(x[0], x[1], color='red', s=100, label='Output = 0')

# Decision boundary: w1*x + w2*y + b = 0
x = np.linspace(-0.5, 1.5, 100)
y = (-(weights[0] * x + bias)) / weights[1]
plt.plot(x, y, 'b-', label='Decision Boundary')

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("NOR Gate Decision Boundary")
plt.grid(True)

# Remove duplicate labels
handles, labels = plt.gca().get_legend_handles_labels()
unique = dict(zip(labels, handles))
plt.legend(unique.values(), unique.keys())

plt.show()
```
