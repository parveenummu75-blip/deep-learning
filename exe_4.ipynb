```python
import numpy as np

def step(x):
    return 1 if x >= 0 else 0

def perceptron(x, w, b):
    net = np.dot(x,w)+b
    return step(net)

# AND Gate
X = np.array([[0,0],[0,1],[1,0],[1,1]])

and_weights = np.array([1,1])
bias = -1.5

print("AND Gate")
for x in X:
    print(x, perceptron(x,and_weights,bias))


# OR Gate
or_weights=np.array([1,1])
bias=-0.5

print("\nOR Gate")
for x in X:
    print(x, perceptron(x,or_weights,bias))


# NAND Gate
nand_weights=np.array([-1,-1])
bias=1.5

print("\nNAND Gate")
for x in X:
    print(x, perceptron(x,nand_weights,bias))
```
