```python
import numpy as np


X=np.array([[0,0],[0,1],[1,0],[1,1]])
Y=np.array([0,0,0,1])


learning_rates=[0.1,0.5,1.0]


for lr in learning_rates:

    weights=np.zeros(2)
    bias=0


    for epoch in range(100):

        error_count=0

        for x,y in zip(X,Y):

            net=np.dot(x,weights)+bias

            output=1 if net>=0 else 0

            error=y-output

            weights+=lr*error*x
            bias+=lr*error

            error_count+=abs(error)


        if error_count==0:
            break


    print("\nLearning Rate:",lr)
    print("Epoch:",epoch+1)
    print("Weights:",weights)
    print("Accuracy:",100*(4-error_count)/4)
```
