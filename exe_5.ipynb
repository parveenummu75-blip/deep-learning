```python
import numpy as np


def step(x):
    return 1 if x>=0 else 0


def sigmoid(x):
    return 1/(1+np.exp(-x))


def relu(x):
    return max(0,x)



inputs=[-2,-1,0,1,2]


print("Input Step Sigmoid ReLU")

for x in inputs:
    print(x,
          step(x),
          sigmoid(x),
          relu(x))
```
