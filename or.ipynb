```python
# Perceptron implementation for OR Gate

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def predict(self, inputs):
        # Calculate weighted sum
        total = 0
        for i in range(len(inputs)):
            total += inputs[i] * self.weights[i]
        total += self.bias

        # Step activation function
        if total >= 0:
            return 1
        else:
            return 0


# Suitable weights and bias for OR gate
weights = [1, 1]
bias = -0.5

# Create perceptron object
perceptron = Perceptron(weights, bias)

# OR gate input combinations
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

# Display output
print("X1 X2 Output")
for x in inputs:
    output = perceptron.predict(x)
    print(x[0], " ", x[1], "   ", output)
```
