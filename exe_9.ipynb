```python
import numpy as np


X=np.array([[0,0],[0,1],[1,0],[1,1]])
Y=np.array([0,1,1,0])


weights=np.zeros(2)
bias=0
lr=0.1


for epoch in range(20):

    for x,y in zip(X,Y):

        output=1 if np.dot(x,weights)+bias>=0 else 0

        error=y-output

        weights+=lr*error*x
        bias+=lr*error



print("XOR Output")

for x in X:

    print(x,
    1 if np.dot(x,weights)+bias>=0 else 0)
```
