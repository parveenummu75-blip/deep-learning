```python
import numpy as np


X=np.array([[0,0],[0,1],[1,0],[1,1]])
Y=np.array([0,0,0,1])


weights=np.zeros(2)
bias=0

lr=0.2


for epoch in range(1,21):

    total_error=0

    for x,y in zip(X,Y):

        net=np.dot(x,weights)+bias

        output=1 if net>=0 else 0

        error=y-output

        weights += lr*error*x
        bias += lr*error

        total_error += abs(error)


    print("Epoch",epoch,
          "Error",total_error)


    if total_error==0:
        print("Converged at epoch",epoch)
        break
```
